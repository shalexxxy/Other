```python
import pandas as pd
```

```python
data1 = pd.read_excel('sample1.xlsx')
data2 = pd.read_excel('sample2.xlsx')
data3 = pd.read_excel('sample3.xlsx')
data = pd.concat([data1,data2,data3])
```

```python
data = pd.read_excel('sample.xlsx')
```

```python
from sentence_transformers import SentenceTransformer
#danasone/bart-small-ru-en
# 1. Load a pretrained Sentence Transformer model
model = SentenceTransformer("all-mpnet-base-v2")

# The sentences to encode
sentences = list(data['Заголовок'])

# 2. Calculate embeddings by calling model.encode()
embeddings = model.encode(sentences)
print(embeddings.shape)
# [3, 384]

# 3. Calculate the embedding similarities
similarities = model.similarity(embeddings, embeddings,)
print(similarities)
# tensor([[1.0000, 0.6660, 0.1046],
#         [0.6660, 1.0000, 0.1411],
#         [0.1046, 0.1411, 1.0000]])
```

```python
similarities.fill_diagonal_(0)
```

```python
import torch
res = []
for i in range(data.shape[0]):
  am = torch.topk(torch.cat((similarities[i,:i],similarities[i, i+1:])), 5)
  res.append(am)
```

```python
import torch
res = []
for i in range(data.shape[0]):
  am = similarities[i].argmax()
  res.append(am)
```
